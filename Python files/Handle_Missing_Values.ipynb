```python
import pandas as pd
import os

dataset_path = r"C:\Final project\Dataset"
```

CUSTOMERS


```python
print("=" * 60)
print("  CUSTOMERS.CSV")
print("=" * 60)
 
customers_path = os.path.join(dataset_path, "Customers.csv")
customers = pd.read_csv(customers_path)
 
customers["last_name"] = customers["last_name"].fillna("Unknown") 
print(f"  last_name → filled 1 NaN value with 'Unknown'")
 
```

PRODUCTS 


```python
print("\n" + "=" * 60)
print("  PRODUCTS.CSV")
print("=" * 60)
 
products_path = os.path.join(dataset_path, "Products.csv")
products = pd.read_csv(products_path)
 
products["recyclable"] = products["recyclable"].fillna(0)
products["low_fat"]    = products["low_fat"].fillna(0)
 
products["recyclable"] = products["recyclable"].astype(int)
products["low_fat"]    = products["low_fat"].astype(int)
 
print(f"  recyclable → filled {687} NaN values with 0")
print(f"  low_fat    → filled {1008} NaN values with 0")
```

SAVE CLEAN FILES

```python
output_path = r"C:\Final project\Dataset"

customers.to_csv(os.path.join(output_path, "Customers.csv"), index=False)
products.to_csv(os.path.join(output_path,  "Products.csv"),  index=False)

print("\n" + "=" * 60)
print("  Clean files saved successfully!")
print("     → Customers.csv")
print("     → Products.csv")
print("=" * 60)
```
