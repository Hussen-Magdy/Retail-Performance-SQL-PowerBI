Detect Missing Values


```python
import pandas as pd
import os
```

```python
dataset_path = r"C:\Final project\Dataset"

files = [
    "Customers.csv",
    "Products.csv",
    "Stores.csv",
    "Region.csv",
    "Calendar.csv",
    "Sales 1997.csv",
    "Sales 1998.csv",
    "Returns.csv",
]
```

```python

print("=" * 60)
print("         MISSING VALUES REPORT - ALL FILES")
print("=" * 60)

for file in files:
    filepath = os.path.join(dataset_path, file)
    df = pd.read_csv(filepath)

    total_rows = len(df)
    missing = df.isnull().sum()
    missing = missing[missing > 0]  

    print(f"\n File: {file}  ({total_rows:,} rows)")
    print("-" * 60)

    if missing.empty:
        print(" No missing values found.")
    else:
        print(f"   {'Column':<35} {'Missing':>8}  {'% of Total':>10}")
        print(f"   {'-'*35}  {'-'*8}  {'-'*10}")
        for col, count in missing.items():
            pct = (count / total_rows) * 100
            print(f"   {col:<35} {count:>8,}  {pct:>9.2f}%")

print("\n" + "=" * 60)
print("                     END OF REPORT")
print("=" * 60)
```
